Explorar los datos

```python
# Importar bibliotecas para explorar datos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importar Datos
data = pd.read_csv('data/ensanut2022_muestras.csv', sep=';')
```

```python
data.head()
```

```python
# Rango por variable
# Tipo de dato de cada variable
data.info()

# Separar variables categoricas y numericas
categorical_vars = data.select_dtypes(include=['object', 'category']).columns
numerical_vars = data.select_dtypes(include=['number']).columns
print("Categorical Variables:", categorical_vars)
print("Numerical Variables:", numerical_vars)

df_cat = data[categorical_vars]
df_num = data[numerical_vars]
```

```python

for col in df_cat.columns:
    print(f"Value counts for {col}:")
    print(df_cat[col].value_counts())
    print("\n")

for col in df_num.columns:
    print(f"Descriptive statistics for {col}:")
    print(df_num[col].describe())
    print("\n")
```

LIMPIEZA DE DATOS SIN CONOCIMIENTO MEDICO

```python
# Importar diccionario de datos
dict_data = pd.read_excel('data/ensanut2022_muestras-diccionario.xlsx')

# Eliminar las primeras 18 filas que no contienen información relevante
dict_data = dict_data.iloc[16:].reset_index(drop=True)
dict_data.head()

#Renombrar columnas
dict_data.columns = ['Variable', 'Posicion', 'Etiqueta', 'Nivel de Medicion', 'Ancho de columna']
dict_data.head()
```

```python
# Mostrar etiquetas de las variables categoricas
for var in categorical_vars:
    etiquetas = dict_data[dict_data['Variable'] == var]['Etiqueta'].values
    if len(etiquetas) > 0:
        print(f"{var}: {etiquetas[0]}")
    else:
        print(f"No se encontraron etiquetas para {var}")
```

```python
# Lista de variables cuyo nombre contiene "valor"
Cat_vars = dict_data.loc[dict_data['Variable'].str.contains("valor", case=False, na=False), 'Variable'].tolist()

# Añadir variables adicionales y eliminar duplicados
Cat_vars = list(set(Cat_vars + ['FOLIO_INT','ponde_v', 'ponde_m12a49', 'ponde_m0a11', 'upm', 'MATRIZ_SUEROMICRO', 'MATRIZ_SUERO']))
```

```python
Cat_vars
```

```python
# Lo mismo con las numericas
# Mostrar etiquetas de las variables numericas
for var in numerical_vars:
    etiquetas = dict_data[dict_data['Variable'] == var]['Etiqueta'].values
    if len(etiquetas) > 0:
        print(f"{var}: {etiquetas[0]}")
    else:
        print(f"No se encontraron etiquetas para {var}")
```

```python
#Añadir variables relevantes a la lista de numericas Sexo, Edad, Matriz_Suero, MAtriz_sangre
Num_vars = ['h0302', 'h0303', 'MATRIZ_SUERO', 'MATRIZ_SANGRE', 'estrato', 'region', 'est_sel']
```
