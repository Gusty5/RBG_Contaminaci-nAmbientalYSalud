```python
import unidecode
```

Se quiere:

1. Quitar los numeros del principio
2. Convertir todo a minusculas
3. Quitar los acentos (con unidecode)

```python
data_selected['desc_ent'] = (
    data_selected['desc_ent']
    .str[3:]          
    .str.lower()             
    .apply(lambda x: unidecode.unidecode(x))
)
```

```python
data_selected['desc_mun'] = (
    data_selected['desc_mun']
    .str[4:]          
    .str.lower()             
    .apply(lambda x: unidecode.unidecode(x))
)
```

```python
data_selected = data_selected.rename(columns={
    "desc_ent": "Entidad",
    "desc_mun": "Municipio"
})
data_selected[['Entidad', 'Municipio']].head()
```

```python
# Move the two ID columns all the way to the right

cols = list(data_selected.columns)
# Take everything except the two you want to move
cols = [c for c in cols if c not in ['Entidad', 'Municipio']] 
# Put them at the end
cols = cols + ['Entidad', 'Municipio']
# Reorder dataframe
data_selected = data_selected[cols]

data_selected.head()
```

```python
ent_mun = pd.read_csv('../data_util/Emisiones_municipio.csv')
ent_mun.head()
```

```python
ent_mun['Entidad'] = (
    ent_mun['Entidad']        
    .str.lower()             
    .apply(lambda x: unidecode.unidecode(x))
)
ent_mun['Municipio'] = (
    ent_mun['Municipio']        
    .str.lower()             
    .apply(lambda x: unidecode.unidecode(x))
)

ent_mun[['Entidad', 'Municipio']].head()
```

```python
data_merged = pd.merge(data_selected, ent_mun, on=['Entidad', 'Municipio'], how="left")
data_merged
```

```python
data_merged.info()
```

```python
cols = data_merged.columns[1:20]

data_merged[cols] = (
    data_merged[cols]
    .replace(r'^\s*$', np.nan, regex=True)
    .replace(",", ".", regex=True)          
    .astype(float)                          
)
```
